```python
#Erkek ve kadın arasındaki hız farklarını 50km ve 50 mide inceleyen sorgu.
```

```python
df3.groupby(["Etkinlik_uzunluğu","Atlet_cinsiyeti"])["Atletin_ortalama_hızı"].mean()
```

```python
#Hangi yaş 50 m yarışında daha iyi?(20 yarışa girmiş en az) 15 adedi göster.
```

```python
df3.query('Etkinlik_uzunluğu == "50mi"').groupby("Atlet_yaşı")["Atletin_ortalama_hızı"].agg(["mean","count"]).sort_values("mean",ascending=False).query("count>19").head(15)
```

```python
#Hangi yaş 50 m yarışında daha kötü?(20 yarışa girmiş en az) 15 adedi göster.
df3.query('Etkinlik_uzunluğu == "50mi"').groupby("Atlet_yaşı")["Atletin_ortalama_hızı"].agg(["mean","count"]).sort_values("mean",ascending=True).query("count>19").head(15)
```

```python
#Yarış sezonlarını aylara ayırıyoruz. Ve hangi ayın ne sırada olduğuna bakarak mevsimsel hızı karşılaştırıyoruz.
```

```python
df3["Yarıs_ayı"]=df3["Etkinlik_tarihi"].str.split(".").str.get(1).astype(int) # Ay kısmını almamızı sağladı.
```

```python
df3["Yarış_sezonu"] = df3["Yarıs_ayı"].apply(lambda x: "Kış" if x > 11 else "Sonbahar" if x >8 else "Yaz" if x > 5 else "İlkBahar" if x > 2 else "Kış") #Ay numarasına göre mevsimleri seçtik.
```

```python
df3.groupby("Yarış_sezonu")["Atletin_ortalama_hızı"].agg(["mean","count"]).sort_values("mean",ascending = False) # Mevsimsel ortalama hız karşılaştırması.
```
